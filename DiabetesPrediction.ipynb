```python
import pandas as pd
```

```python
df=pd.read_csv('Diabetes_and_LifeStyle_Dataset .csv')
df
```

```python
df.isna().sum()
```

```python
df.columns
```

#### Dropped unwanted columns

```python
print(df.drop(['diabetes_risk_score','diabetes_stage','ethnicity','education_level','income_level','employment_status'],axis=1,inplace=True))
```

```python
df.columns
```

### Feature Engineering

```python
#Mapping
df['gender']=df['gender'].map({'Male':0,'Female':1})
```

```python
df['gender'].unique()
```

```python
df.isna().sum()
```

```python
a=df['gender'].ffill()
print(df['gender'].fillna(a,inplace=True))
```

```python
df.isna().sum()
```

```python
df['smoking_status'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['smoking_status'])
df
```

```python
df.isna().sum()
```

```python
#Mapping
df['smoking_status_Current']=df['smoking_status_Current'].map({False:0,True:1})
df['smoking_status_Former']=df['smoking_status_Former'].map({False:0,True:1})
df['smoking_status_Never']=df['smoking_status_Never'].map({False:0,True:1})
```

```python
df['smoking_status_Current'].value_counts()
```

```python
print(df['smoking_status_Never'].unique())
```

```python
df.dtypes
```

```python
#Labeling data
x=df.drop(['diagnosed_diabetes'],axis=1)
```

```python
y=df['diagnosed_diabetes']
y.head()
```

### Split Train Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
df['diagnosed_diabetes'].value_counts()
```

### Sampling and Model Selection

```python
from imblearn.over_sampling import SMOTE
smt=SMOTE()
x_train,y_train=smt.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
#Model training and testing
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
lr_acc=accuracy_score(y_test,y_pred)
print(lr_acc)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(7,7))
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues',cbar=False,
            xticklabels=['Predicted Class 0','Predicted Class 1'],
            yticklabels=['Actual Class 0','Actual Class 1'])  #annot- used for display values inside graph #fmt- used for showing int value
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
dtc_pred=dtc.predict(x_test)
dtc_pred
```

```python
dtc_acc=accuracy_score(y_test,dtc_pred)
print(dtc_acc)
```
